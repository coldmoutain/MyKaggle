```python
full = train.append( test , ignore_index = True )
print ('合并后的数据集:',full.shape)
```

```python
full['date']=full.datetime.map( lambda x :x.split()[0] )
full['month']=full.datetime.map( lambda datetime :datetime.split()[0].split('-')[1] )
full['hour']=full.datetime.map( lambda datetime :datetime.split()[1].split(':')[0] )
full['weekday']=full.date.map( lambda date :calendar.day_name[datetime.strptime(date,"%Y-%m-%d").weekday()])
```

```python
full['season'] = full['season'].map({1:'春天',2:'夏天',3:'秋天',4:'冬天'})
full['workingday'] = full['workingday'].map({1:'工作日',0:'非工作日'})
full['weather'] = full['weather'].map({1:'晴天/多云',2:'雾天/阴天',3:'小雨/雪',4:'大雨/雪/雾'})
```

```python
categoryVariableList = ["hour","weekday","month","season","weather","holiday","workingday"]
for var in categoryVariableList:
    full[var] = full[var].astype("category")
```

```python
full  = full.drop(["datetime"],axis=1)
full.info()
```

```python
def rmsle(y, y_,convertExp=True):
    if convertExp:
        y = np.exp(y),
        y_ = np.exp(y_)
    log1 = np.nan_to_num(np.array([np.log(v + 1) for v in y]))
    log2 = np.nan_to_num(np.array([np.log(v + 1) for v in y_]))
    calc = (log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfModel = RandomForestRegressor(n_estimators=100)
yLabelsLog = np.log1p(yLabels) 
rfModel.fit(dataTrain,yLabelsLog)
preds = rfModel.predict(X= dataTrain)
print ("RMSLE Value For Random Forest: ",rmsle(np.exp(yLabelsLog),np.exp(preds),False)) 
```
